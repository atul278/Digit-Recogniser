```python
a=pd.get_dummies(y_trainn)
```

```python
y_train=pd.concat([y_trainn,a],axis=1)
```

```python
y_train=y_train.drop("label",axis=1)
```

```python
x_train=x_train.reshape(x_train.shape[0],28,28,1).astype("float32")
x_test=x_test.reshape(x_test.shape[0],28,28,1).astype("float32")
```

```python
from keras.layers import Dense,Convolution2D,MaxPool2D,Dropout,Flatten
```

```python
from keras.models import Sequential
```

```python
model=Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(28,28,1),activation="relu"))
```

```python
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(128,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(128,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(10,activation="softmax"))
```

```python
model.compile("adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,validation_split=0.2,epochs=25)
```

```python
y_pre1=model.predict(x_test)
```
